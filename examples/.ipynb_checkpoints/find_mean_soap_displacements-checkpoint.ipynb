```python
# Christian Jorgensen
# May 2
# Finding mean displacements of atoms in SOAP space for molecules undergoing crystallization
# Those with the largest displacement are the most interesting to simulate

import numpy as np
from rascaline import SoapPowerSpectrum
from scipy.spatial.transform import Rotation as R
from tqdm.auto import tqdm
from metatensor import Labels, TensorBlock, TensorMap
from itertools import product
from representation_helpers_2023 import get_soap
from ase.io import read, write
from ase.visualize import view
import json
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.linalg import norm
from ase.visualize.plot import plot_atoms
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import colormaps
from sklearn.preprocessing import StandardScaler
from skmatter.preprocessing import StandardFlexibleScaler
import chemiscope
```

```python
def sort_molecule_by_aiim(feats, frame):
    """Sort a molecule's soap vectors according to its aiim

    Args:
        feats (np.ndarray): soap featurization
        frame (Atoms): Frame containing aiim

    Returns:
        np.ndarray: sorted soap vectors
    """
    unordered_pairs = list(zip(feats, frame.arrays['atom_index_in_mol'])) #Pair meaning atom soap + aiim
    sorted_pairs = sorted(unordered_pairs, key = lambda x: x[1]) #Sort so that aiim is ordered

    sorted_mol_feats = np.array([x[0] for x in sorted_pairs])
    
    return sorted_mol_feats

def check_atom_alignment(crystal_frame, mol_frame):
    """Check if the atom indices between two molecules are the same

    Args:
        crystal_frame (Atoms): Frame of molecule in the crystal
        mol_frame (Atoms): Frame of molecule in the gas phase
    """
    unordered_pairs = list(zip(crystal_frame.arrays['numbers'], crystal_frame.arrays['atom_index_in_mol']))
    sorted_pairs = sorted(unordered_pairs, key = lambda x: x[1]) #Sort so that aiim is ordered
    sorted_list = [x[0] for x in sorted_pairs]
    
    isAligned = sorted_list == list(mol_frame.arrays['numbers'])
    return isAligned
    
def get_molecule_within_crystal(crystal, molecule_crystal_idx):
    """Gets a molecule within a crystal

    Args:
        crystal (Atoms): Frame of the crystal structure
        molecule_crystal_idx (int): Index of the molecule of interest in the crystal structure, should be in crystal.arrays['unique_mol_indices']

    Returns:
        frame: frame of the molecule in the crystal
        frame: frame of the crystal
    """

    molecule_in_crystal = crystal[crystal.arrays['unique_mol_indices'] == molecule_crystal_idx]
    return molecule_in_crystal, crystal

def compute_avg_atom_displacement(feats1, feats2, return_disp_per_atom = False):
    """Compute the average displacement of the atoms between two frames

    Args:
        feats1 (ndarray): Molecule 1 featurization in soap space
        feats2 (ndarray): Molecule 2 featurization in soap space

    Returns:
        float: Average displacement of the atoms in the molecule
    """
    displacement_matrix = feats1 - feats2
    displacement_per_atom = norm(displacement_matrix, axis = 1)
    mol_displacement = np.sum(displacement_per_atom)
    avg_displacement = mol_displacement/len(displacement_per_atom)
    
    if return_disp_per_atom:
        return avg_displacement, displacement_per_atom
    
    return avg_displacement

def pad_with_zeros(array1, array2):
    rows1, cols1 = array1.shape
    rows2, cols2 = array2.shape
    
    # Ensure both arrays have the same number of rows
    if rows1 != rows2:
        raise ValueError("Both arrays must have the same number of rows")

    # Determine the maximum number of columns
    max_cols = max(cols1, cols2)
    
    # Create new arrays with the same number of rows and the maximum number of columns
    new_array1 = np.zeros((rows1, max_cols))
    new_array2 = np.zeros((rows2, max_cols))
    
    # Copy the contents of the original arrays into the new arrays
    new_array1[:, :cols1] = array1
    new_array2[:, :cols2] = array2
    
    return new_array1, new_array2
```

```python
# Import molecular crystals dataset
#Fluorinated
#crystal_frames = read("xyz_files/new_all_crystals.xyz", ":")
#mol_frames = read("xyz_files/new_all_relaxed_molecules.xyz", ":")

#Non-Fluorinated
crystal_frames = read("xyz_files/all_crystals.xyz", ":") #37 is a co-crystal
mol_frames = read("xyz_files/all_relaxed_molecules.xyz", ":")
```

```python
#Preprocess the frames
for frame in crystal_frames:
    frame.pbc = True
    frame.wrap(eps=1e-10)

for frame in mol_frames:
    frame.pbc = True
    frame.wrap(eps=1e-10)
    
#Kick out hydrogens, these contribute a lot of noise to the overall displacement
crystal_frames = [frame[frame.numbers != 1] for frame in crystal_frames]
mol_frames = [frame[frame.numbers != 1] for frame in mol_frames]
```

```python
hyperparameters = json.load(open("soap_hypers.json"))
```

crystal.arrays is a dictionary containing:

'numbers': Atomic numbers of each atom in the crystal

'positions': Cartesian coordinates of each atom in the crystal

'atom_index_in_mol': Index of atom in its molecule

'mol_indices': Index of molecule in the DATASET corresponding to the atom in the crystal

'umol_indices': Index of molecule in the CRYSTAL corresponding to the atom in the crystal


## STEPS:

1. Split atoms in crystal into constituent molecules using 'umol_indices' --> set of molecules in the crystal

2. Map each molecule in the crystal to their counterparts in the molecules dataset using 'mol_indices', check that the atoms match up

3. SOAP using molecules generated in (1), do not need to consider coordinates because of invariance

4. Sort the atoms within each molecule using 'atom_index_in_mol'

5. Compute displacement matrix by subtracting SOAPs and computing norm across atoms then across the molecule, standardize by # of atoms


```python
#Break up crystals into their constituent molecules

mol_in_crystal = [] #Frames of individual molecules within all crystals

for crystal_idx, crystal in enumerate(crystal_frames):
    unique_molecules_idx = np.unique(crystal.arrays['unique_mol_indices'])
    
    for unique_mol_idx in unique_molecules_idx:
        mol_in_crystal_frame, _ = get_molecule_within_crystal(crystal, unique_mol_idx)
        mol_in_crystal.append((mol_in_crystal_frame, crystal_idx, unique_mol_idx))

mol_in_crystal_frames = [mol[0] for mol in mol_in_crystal]
crystal_idxs = [mol[1] for mol in mol_in_crystal]
umol_idxs = [mol[2] for mol in mol_in_crystal]
mol_lens = [len(mol) for mol in mol_in_crystal]
```

```python
#Put all the frames together and make SOAP vectors for all of them simultaneously
feats = np.array(get_soap(mol_frames + mol_in_crystal_frames, hyperparameters))
```

```python
total_num_atoms_in_mol_dataset = sum([len(mol) for mol in mol_frames])
total_num_atoms_in_crystal_dataset = sum(len(crystal) for crystal in crystal_frames)
total_num_atoms_in_mol_dataset + total_num_atoms_in_crystal_dataset
```

```python
#Split up SOAPs into those corresponsing with molecule dataset vs crystal dataset
mol_feats = feats[0:total_num_atoms_in_mol_dataset]
mol_in_crystal_feats_unsplit = feats[total_num_atoms_in_mol_dataset:total_num_atoms_in_mol_dataset + total_num_atoms_in_crystal_dataset]
assert len(mol_feats) + len(mol_in_crystal_feats_unsplit) == total_num_atoms_in_mol_dataset + total_num_atoms_in_crystal_dataset
```

```python
#Split up by individual molecule
idx = 0
mol_in_crystal_feats_split = []
for i, mol in enumerate(mol_in_crystal_frames):
    mol_in_crystal_feats_split.append(mol_in_crystal_feats_unsplit[idx:idx + len(mol)])
    idx += len(mol)

idx = 0
mol_feats_by_mol = []
for i, mol in enumerate(mol_frames):
    mol_feats_by_mol.append(mol_feats[idx:idx + len(mol)])
    idx += len(mol)
# mol_by_crystal_feats = []
# idx = 0
# for mol in 
```

```python
#Group molecules by their corresponding crystal
idx = 0
mol_by_crystal_feats = [[] for i in crystal_frames]
for i, mol in enumerate(mol_in_crystal_feats_split):
    current_crystal_idx = crystal_idxs[i]
    mol_by_crystal_feats[current_crystal_idx].append(mol)
```

```python
results = []
j = 0

for crystal_idx, crystal in enumerate(crystal_frames):
    #STEP 1: Split atoms in crystal into constituent molecules using 'umol_indices'
    
    #Get indices to slice by individual molecules in the crystal with
    unique_molecules_idx = np.unique(crystal.arrays['unique_mol_indices'])

    #Break crystal into molecules using idx
    for i in unique_molecules_idx:
        mol_in_crystal_frame, _ = get_molecule_within_crystal(crystal, i)
    
    #STEP 2: Map molecule to molecule in the molecules dataset using 'mol_indices' and check that they are aligned properly
    
        mol_frames_idx = mol_in_crystal_frame.arrays['mol_indices'][0] #Index of corresponding molecule in mol_frames
        mol_in_gas_frame = mol_frames[mol_frames_idx]
        isAligned = check_atom_alignment(mol_in_crystal_frame, mol_in_gas_frame)
        
        if not isAligned:
            print(f"Molecule {i} in Crystal {crystal_idx} is not aligned with Molecule {mol_frames_idx}")
            continue
            
    #STEP 3: SOAP
    
        # mol_in_crystal_feats = np.array(get_soap(mol_in_crystal_frame, hyperparameters))
        # mol_in_gas_feats = np.array(get_soap(mol_in_gas_frame, hyperparameters))
        mol_in_crystal_feats = mol_by_crystal_feats[crystal_idx][i]
        mol_in_gas_feats = mol_feats_by_mol[mol_frames_idx]
        
        if mol_in_crystal_feats.shape != mol_in_gas_feats.shape:
            print(f"Molecule {i} in Crystal {crystal_idx} does not form SOAP vectors of the same shape as Molecule {mol_frames_idx}")
            mol_in_crystal_feats, mol_in_gas_feats = pad_with_zeros(mol_in_crystal_feats, mol_in_gas_feats)
            
    #STEP 4: Sort using 'atom_index_in_mol'
    
        mol_in_crystal_feats_sorted = sort_molecule_by_aiim(mol_in_crystal_feats, mol_in_crystal_frame)
        
    # #STEP 5: Compute displacement and save.

        avg_displacement, atom_displacement = compute_avg_atom_displacement(mol_in_crystal_feats_sorted, mol_in_gas_feats, return_disp_per_atom = True)
        
        results.append({'crystal_dataset_idx': crystal_idx,
                          'CCDC_ID': mol_in_crystal_frame.info['CCDC_ID'],
                          'molecule_dataset_idx': mol_frames_idx,
                          'molecule_crystal_idx': i,
                          'num_atoms': len(mol_in_gas_frame), 
                          'avg_displacement': avg_displacement,
                          'total_displacement': avg_displacement*len(mol_in_gas_frame),
                          'per_atom_displacement': atom_displacement
                          })
        j += 1
        
        if j % 100 == 0:
            print(j)
```
